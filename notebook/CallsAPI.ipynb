# Metodo GET

## Endpoint 1 Solicitud de Usuarios

```python
url = "http://localhost:5000/User_group"
dato = {'Type_Chat':'Grupo', 'Name_Chat':'recuvenza', 'Collection': 'Trabajo'}
frases = requests.get(url, data=dato)
frases.json()
```
